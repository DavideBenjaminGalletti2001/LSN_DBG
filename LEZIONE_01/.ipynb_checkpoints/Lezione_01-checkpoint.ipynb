<h2><font color = 'LimeGreen'>ESERCIZIO 1.1: Uniformita' e DataBlocking</font></h2>

In termini semplici, il metodo del data blocking indica essenzialmente una procedura finalizzata ad ottenere campionamenti statisticamente indipendenti in cui sono applicabili le ipotesi del teorema del limite centrale(TLC), la quale consiste nel raggruppare le singole misure-sia che esse siano dirette, che indirette- in blocchi, in ciascuno dei quali si calcola poi media $\langle x \rangle = \frac{\sum_{1}^{N} x_{i}}{N}$ . Tali valor medi, per blocchi sufficientemente grandi in modo tale da abbattere le correlazioni e ridurle a meri effetti di superficie, rappresentano i singoli esperimenti su cui poi applicare il TLC e stimare l'incertezza.
Nello specifico, per stimare in modo fine una corretta dimensione dei blocchi, il procedimento da adottare prevede di analizzare l'autocorrellazione tra misure successive e regolare conseguentemente la dimenzione affinche' questa sia maggiore del "tempo" di correlazione, ove per tempo si intende tempo macchina e quindi il numero di valutazioni.

Il metodo del data blocking Ã¨ anche utile per osservare l'andamento progressivo delle misure che si stanno effettuando nella simulazione, fornendo informazioni supplementari al semplice risultato finale con rispettiva incertezza.

<h3><font color = 'LightGreen'> Test $\chi ^2$</font></h3>

I test realizzati finora si basano sull'ipotesi fondamentale che i numeri estratti dal generatore di numeri pseudo-casuali siano veramente distribuiti uniformemente nell'intervallo $[0,1]$. Per verificarlo si ricorre al test del $\chi^2$. Per implementarlo, si divide l'intervallo in tot sottorvalli, nello specifico si e' preso $M = 100$,per poi estrarre $ n = 10^4$ valori e valutare la funzione:
<p style="text-align:center">$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$</p>
dove $n_k$ e' uguale al numero di campionamenti nel sottointervallo i-esimo: $[k*0.01, (k+1)*0.01)$
In media, $(n_i - n/M)^2 \simeq n/M$ e quindi si puo' affermare che il generatore sara' uniforme se $\chi^2 \simeq 100$

```python
# Continuo usando variabili e funzioni della cella precedente

plt.figure(figsize=(10,7), facecolor = "AliceBlue")

chi = np.loadtxt("EX_1.1_C/test.dat", usecols=(1), unpack='true')
plt.plot(x, chi-100, color = "Brown", marker = 'o', mec = 'k', mfc = 'k')
font = {"family":"serif", "color":"ForestGreen", "size":"15"}
plt.xlabel('Iterazioni ', fontdict = font)
plt.ylabel('$\chi^2-100$', fontdict = font)
plt.title("Test $\chi ^2$", fontsize=25, color ="ForestGreen", family="serif", fontweight = "bold")
plt.grid(True, alpha = 0.3)
plt.axhline(y=0, color = 'r', linestyle = '-')

plt.show()

print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
print('MEDIA (atteso 100): ')
print(np.mean(chi))
print('VARIANZA (attesa 200): ')
print(np.var(chi))
print('INCERTEZZA (atteso 14.1): ')
print(np.std(chi))
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
```

<h2><font color = 'LimeGreen'>ESERCIZIO 1.2: TLC e Distribuzioni Stabili</font></h2>

Accertatisi della bontÃ  del generatore di numeri casuali, il secondo esercizio richiede di verificare il Teorema del limite centrale.

Il Teorema dice che la media $A_N = \frac{1}{N}\sum_{i=1}^N x_i$di un numero sufficientemente grande di valori casuali aleatorie indipendenti, purchÃ¨ di varianza finita, tende ad assumere una forma gaussiana.

Nello specifico, si considerano il caso di distribuzione uniforme, lorentziana ed esponenziale.
Per questi casi elementari si puo' usare la tecnica della inversione della cumulativa. Questa tecnica permette di ottenere valori di x in una distribuzione che risulta in y=F(x) estraendo in modo uniforme valori di y in [0,1) ed inserendoli nell'inversa della funzione cumulativa:
<p style="text-align:center"> $F(x) = \int_{0}^{x}p(x')dx'=:y \Rightarrow x= F^{-1}(y)$</p>
Svolgendo banali conti:<br>

1. $p(x) = \lambda \exp(-\lambda x), x\in [0;+\infty] \Rightarrow x=-\frac{1}{\lambda}\ln(1-y)$
2. $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}, x\in [-\infty;+\infty] \Rightarrow x=\Gamma \tan [\pi (y-\frac{1}{2})] +\mu$

```python
plt.figure(figsize = (20,20), facecolor = "AliceBlue")
myTitle = np.array(["uni", "exp", "lorentziana"])
myColor = np.array(["DarkCyan", "CornFlowerBlue", "Cyan", "LightSkyBlue"])
N = [1,2,10,100]
conta = 1
for i in myTitle:
    for j in range(0,len(N)):
        plt.subplot(3,4,conta)
        y = np.loadtxt("EX_1.2/"+i+str(N[j])+".dat", usecols = (1), unpack = "true")
        plt.plot(x,y, marker = ".", label = "N = " + str(N[j]), color = myColor[j])
        plt.legend(fontsize = 12)
        plt.tick_params(axis='both', which='major', labelsize=13.5)
        conta+=1
        
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
plt.show()        
```

Nei primi due casi, dado uniforme ed esponenziale, Ã¨ manifesta la convergenza alla distribuzione normale. Gli istogrammi per la distribuzone lorentziana, invece, rimangono sostanzialmente immutati al variare della grandezza di popolazione su cui stiamo mediando: questo risultato Ã¨ compatibile con il Teorema del Limite Centrale il quale, infatti, vale solo sotto l'assunzione che la varianza della distribuzione di partenza sia finita mentre la Lorentziana ha varianza infinita. Altrettando interessante e' osservare come la distribuzione Lorentziana tenda a se stessa trattandosi, come la gaussiana, di una distribuzione stabile.

<h2><font color = 'LimeGreen'>ESERCIZIO 1.3: Buffon e Calcolo di $\pi$</font></h2>

Se lanciamo casualmente un ago di lunghezza $L$ su un piano orizzontale dove sono posizionate delle linee parallele e distanti tra loro di una distanza $d > L$, ripetendo tante volte l'esperimento e' possibile trovare la probabilita' che l'ago intersechi una delle linee. Buffon poi intui' come a partire da questa considerazione si potessi ricavare la costante di $\pi$. Contando il numero di volte $M$ che la'go interseca una linea, si trova:
$\pi = \lim_{N \to \infty}\frac{2LN}{Md}$. Cosi' facendo si ha modo di ricavare la costante di pigreco.
In questo esercizio l'obiettiva e' realizzare un algoritmo che riproduca il procedimento descritto originariamente da Buffon.

```python
N = 100
x = np.arange(N)+1      
plt.figure(figsize=(10,7), )

ave, err = np.loadtxt("EX_1.3/risultati.dat", usecols=(1,2), delimiter=' ', unpack='true')

plt.title("CALCOLO: $\pi$", fontsize=25, color ="ForestGreen", family="serif", fontweight = "bold")
plt.errorbar(x, ave, yerr = err, color = "DarkCyan", ecolor = "LightSkyBlue")
font = {"family":"serif", "color":"ForestGreen", "size":"15"}
plt.xlabel('Block' , fontdict = font)
plt.ylabel(r'$\langle \pi \rangle$', fontdict = font)
plt.grid(True, alpha = 0.3)
plt.axhline(y=math.pi, color = 'r', linestyle = '-', label = "$\pi$")
plt.legend()

plt.show()

print('PI medio (atteso '+str(np.pi)+'): ')
print(ave[-1])
print('\nErrore statistico : ')
print(err[-1])
```
