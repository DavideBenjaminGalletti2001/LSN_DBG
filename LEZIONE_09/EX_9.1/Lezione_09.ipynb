<h1><font color='ForestGreen'>ESERCITAZIONE 9: Ottimizzazione e Algoritmi Genetici </font> </h1>

<h3><font color = 'LightGreen'>Algoritmi Genetici</font></h3>

In questa esercitazione l'obiettivo e' risolvere il problema del commesso viaggiatore servendosi dei cosiddetti algoritimi genetici, una tecnica di 
ottimizzazione che prende ispirazione dai processi biologici delle specie animali e dai fenomeni di mutazione genetica. In sistesi, tale tecnica si puo' definire:
<lo>
    <li>Gradient-Free </li>
    <li>Stochastic </li>
    <li>Population Based</li>
    <li>Global Saerching</li>
</lo>
Per essere piu' chiari, i GAs per lavorare soddisfano alcune caratteristiche: generalmente ricorrono ad una codifica dei parametri, piuttusto che usare i parametri effettivi; sfruttano una modalita' di ricerca in cui viene manipolata ed elaborata una popolazione di soluzioni candidate; estraggono informazione in modo diretto, cioe' senza far uso di gradiente; contempleano un elemento di stocasticita', per cui hanno proprieta' di transizione
probabilistiche.Come codifica, vi sono molteplici scelte; per esemplificare e' possibile considerare una stringa binaria, un numero reale, strutture dati oppure pure elementi di programmi.

in termini schematici, il funzionamento di un algoritmo genetico consta dei seguenti punti fondamentali:

1. Si parte da una popolazione di n-cromosomi, cioe' soluzioni candidate codificate: la scelta piu' semplice e conveniente spesso e' quella di stringhe di bit.
2. Si calcola il fitness per ogni cromosoma della popolazione: tale quantita' e' una misura della capacita' di ottimizzazione ed e' tanto piu' basso
   quanto maggiore e' il fitness.
3. Si ripetono i seguenti step finche' non si decide di stoppare l'evoluzione
   <p style = "text-align:left">
       - Si seleziona una coppia di cromosomi genitori con probabilita' proporzionale al fitness, cioe' si implementa un metodo
           per simulare i processi di selezione naturale:  una sorta di rulette truccata. L'estrazione dell'indice segue in particolare la legge:
                 $j= int(M \times r^p)+1$<br>
      - Con probabilita' $p_c$ si prendera' questa coppia e si produrra incrocio a punto fisso, tipicamente a punto singolo.
           tale procedura va a rapprasentare il fenomeno di Crossover.<br>
    - Si effettua mutazione di queste nuovi o vecchi inidividui, il che riproduce il concetto di alterazione genetico.<br>
   </p>
5. Si sostitusice la popolazione di partenza cn una nuova, generalmente assunta della medesima dimensione.
6. Si ripete il processo generando via via una generazione dopo l'altra, cioe' una popolazione di cromosomi-individui-da cui selezionare l'ottimo valutandoli sulla funzione costo.

Un GAs, in breve, quindi prende una popolazione-insieme- di soluzioni candidate -cromosomi- del problema, quali possono essere per il problema del commesso viaggiatore nient'altro che i possibili percorsi, per ogni elemento di tale popolazione di va a valutare la funzione perdita per imporre la selezine naturale e generare cosi' una generazione di figli, cioe' una nuova popolazione cui si ripete la selezione di cui sopra in funzione dell'adattabilita'. Insieme poi alla selezione naturale, i cromosomi che vanno a costituire la popolazion epossono a loro volta subire alterazioni genetiche: anzitutto crossover, dopodiche' mutazioni genetiche di varia natura.

<h3><font color = 'LightGreen'>Intro: Problema del Commesso Viaggiatore TSP</font></h3>

Come osservato, quanto ci si propone e' utilizzare gli algoritmi genetici per ottimizzare il problema del commesso viaggiatore:
<br>
<span style = "color:blue"> il commesso deve visitare una e una sola volta citt√† in posizioni partendo da una e tornandoci, minimizzando il percorso totale </span> 
<br>
Tale percorso, sara' della forma:
<p style = "text-align:center">
$L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|^2
\qquad \text{oppure} \qquad
L^{(1)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|$
</p>

In maniera arbitraria, si e' optato per la prima opzione.

Forse non seguendo il principio della strada piu' semplice e sicura, in questo elaborato si e' codificato il singolo cromosoma-rappresentante il percorso
fatto dal commesso viaggiatore-come un vector-sfruttando la potenza e versatilita' della classe STL in c++ di iteratori ad un opportuno vector di citta', ove queste sono rappresentate a loro volta da una ulteriore classe per distinguere le varie casisitiche: city; city_circ; city_quad...; cioe' citta' distribuite casualmente su un piano; citta' distribuite su circonferenza; citta' distribuite dentro un quadrato...
Ragion per cui, va da se' che la classe TSP andra' ad essere templetizzata.
Un'alternativa, altrettanto valida e forse concettualmente e nella pratica piu' semplice da implementare, sarebbe stata scegliere una codifica a vector di interi: etichettando ogni citta' con un numero, si sarebbe perfettamente potuto ottenere ogni percorso tramite semplice permutazione: procedura che, tramite  semplici modifiche del codice, e' stata implementata nell'esercitazione 10. Tale necessita' e' derivata dal fatto di dover ricorrere, per parallelizzare il programma, alle librerie MPI: d'altra parte, queste supportano, in linea di principio, solo il formato C, il che rende impossibile specificare vector di classi personalizzati. In questo caso addirittura vector di iteratori ad oggetti non primitivi del linguaggio come la classe rappresentante le citta'. Un alternativa, disponibile sul mercato, e' ricorrere alle librerie Boost MPI: una estensione c++ dell'MPI.
