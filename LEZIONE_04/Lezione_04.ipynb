<h3><font color = 'LightGreen'>Intro: Ensamble Microcanonico</font></h3>

In termini generali si intende per ensemble statistico l'insieme di tutti i possibili (micro)stati di un sistema fisico compatibili con un certo macrostato definito dai vincoli sperimentali. Si distinguono: l'ensamble microcanonico, canonico e grancanonico. In questa esercitazione si vogliono applicare tecniche di dinamica molecolare per simulare ensemble microcanonico nei differenti stati d'aggregazione della materia: nello stato solido, liquido e gassoso, ognuna caratterizzata dalla propria rispettiva temperature.

In quanto identificato dall'avere energia, numero di particelle e volume costante, l'ensable microcanonico sara' identificato da un sistema isolato: motivo per cui, giustificato e' anche denominarlo come sistema NVE, la quale sigla indica la costanza di N, V ed E. Il fatto poi che il valore della energia sia fisso conduce al I postulato della meccanica statistica, noto pure come postulato di uguale probabilita' a priori: in parole parole, questo dice che autostati degeneri del medesimo livello di energia sono equiprobabili; equiprobabilita' giustificata dalla radicata convinzione che la naturali non ha particolare esigenza nel preferire stati aventi la medesima energia.

Per l'esercitazione e' stato gia preliminarmente consegnato un codice già funzionante  e compilabile in grado di simulare l'evoluzione dinamica di un sistema di particelle tramite l'algoritmo di Verlet, e di misurarne energia totale per particella , energia potenziale per particella , energia cinetica per particella  e temperatura . Il programma può peraltro salvare le posizioni delle molecole nello spazio 3D, utile per visualizzare il moto del sistema tramite software appositi come Ovito.



<h3><font color = 'LightGreen'>Algoritmo di Verlet</font></h3>

Con l'obiettivo di andare ad integrare le equazioni di Newton,l'algoritmo di Verlet si ottiene andando ad espandere al terzo ordine rispetto a 
$t+\delta t$ e $t-\delta t$ e andando poi a sommare le due equazioni. Dopo alcuni semplici passaggi, si arriva:
<p style = "text-align:center">
$\vec r(t+dt)=2\vec r(t)-\vec r(t-dt)+\vec a(t)dt^2$
</p>
Noto il potenziale di interazione tra coppie di particelle, la cui scelta se vogliamo e' il cuore della dinamica molecolare e dello studio dei sistemi di particelle, la quale funge da una sorta di microscopio computazionale, si puo' andare a calcolare la accelerazione partendo dalla forza agente su ogni particella in ognuna della tre dimensioni ottenuta come meno gradiente del potenziale. Come tipico in un contesto quale materia condendata, l'interazione e' assunta della forma LJ:
<p style = "text-align:center">
$V = 4\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right]$
</p>
Trovate le posizione, si ricavano facilmente le velocita' con un errore dell'ordine $\theta (\delta t^2)$. Piu' esplicitamente, si ha
<p style = "text-align:center">
$\vec{v}(t) = \frac{\vec{r}(t+\delta t) - \vec{r}(t-\delta t)}{2\delta t}$
</p>



<h2><font color = 'LimeGreen'>ESERCIZIO 4.1: Teorema del viriale, calcolo pressione</font></h2>

In questo primo punto dell'esercitazione si e' semplicemente introdotto il calcolo della pressione all'interno del codice consegnato
gia' funzionante di dinamica molecolare. Tale quantita' e' stata determinata sfruttando il teorema del viriale. Eseguendo il conto si arriva a :
<p style = "text-align:center">
$P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $
</p>
Passando ora in unita' ridotte in modo tale sfruttare la legge degli stati corrispondne e avere grandezze piu' adeguate ad essere manipolate da un
calcolatore per evitare problemi di underflow e overflow, si ottiene:
<p style = "text-align:center">
$P^* = \rho^* T^* + 
                \frac{1}{3V^*}
                \left\langle
                    \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 
                        48 \left[   
                            \left(
                                \frac{1}{|\vec{r}_i -\vec{r}_j|}
                            \right)^{12} 
                            -\frac{1}{2}
                            \left(
                                \frac{1}{|\vec{r}_i -\vec{r}_j|}
                            \right)^6 
                           \right] 
                \right\rangle
    = \rho^* T^* + 
                \frac{
                \left\langle
                    W 
                \right\rangle
                }{3V^*} $
</p>

Si osservi:
<p style = "text-align:center">
$P^{*}=P\frac{\sigma^3}{\epsilon},\quad T^*=k_{B}\frac{T}{\epsilon}, \quad V^*=\frac{V}{\epsilon}, \quad\rho^*=\rho\sigma^3, \quad r^*=\frac{r}{\sigma} \quad t^{*} = t \cdot \sqrt{\frac{\epsilon}{\sigma^{2}m}}$
</p>

<h3><font color = 'LightGreen'>Equilibrazione</font></h3>

In quasta prima fase della simulazione il compito e' permettere al sistema di equilibrarsi e raggiungere una condizione che rifletta la termodinamica e i vincoli del problema.In pratica, data una configurazione cristallina, occorre aspettare un tempo necessario affinche' raggiunga la condizione d'equilibrio e si possa impiegare la configurazione finale come input per svolgere la simulazione vera e propria della termodinamica del sistema.Il problema, tuttavia,e' che se partiamo fissando la temperatura al valore target, in breve tempo ci si accorge che questa tende a deviare notevolemente dal valore di input portando ad un risultato sbagliato. Motivo per cui, per non introdurre errori sistematici, la procedura corretta prevede di effettuare molte prove cambiando la temperatura di input e monitorando i valori della temperatura istantanea usando un numero dufficiente di blocchi da uno step: fatto cio', il sistema si sara' stabilizzato al valore di temperatura target e si potra' iniziare a campionare le osservabili di interesse.
I plot qui sotto riportati mostrano proprio l'andamento della temperatura istantanea in funzione del numero di steps.

Trovata poi questa temperatura, basta copiare i file velocity.out e config.out, che riportano rispettivamente velocita' e posizioni delle 108 particelle alla fine della termalizzazione, nella cartella "fase/equilibrio" e nel file di input sostituiamo 1 allo 0 nel campo restart. In questo modo, cambiando i valori del numero di blocchi e di step per effettuare la vera simulazione, il codice ripartira' dalle configurazioni di equilibrio assicurando che il sistema sara' gia' termalizzato.

I dati del problemi, al variare della fase, sono qui sotto riportati:

In generale, si ha: $\sigma = 0.34, \epsilon/k_B = 120, m=39.948, N = 108$

1. solid phase: $\rho^\star = 1.1$, $T^{\star}_{input} = 1.54$, $T^{\star}_{target} = 0.8$,  (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$,$T^{\star}_{input} = 1.2$, $T^{\star}_{target} = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^{\star}_{input} = 0.96$, $T^{\star}_{target} = 1.2$ (cut-off radius: $r_c = 5.0$)

<h3><font color = 'LightGreen'>Termodinamica:</font></h3>

```python
- Energia totale $E=K+U$ <br>
- Energia cinetica  $K_{tot}=\sum_{i=0}^N \frac{1}{2}m(v_x^2+v_y^2+v_z^2)_i$ <br>
- Energia potenziale $U_{tot}=\sum_{i=0}^N U_i$ dove ogni energia potenziale è nella forma di LJ <br>
- Temperatura $T = 2K_{tot}/(3N)$ <br>
```

```python
npart = 108
N = 60
x = np.arange(N)+1    # uso l'ordinamento da 1 a N
```

```python
th_var = ["ekin", "epot", "etot", "pres", "temp"]

titles = [  "Energia Cinetica",
            "Energia Potenziale",
            "Energia Potale",
            "Pressione", 
            "Temperatura",
            "Temperatura"]
colors = ["Brown", "Gold", "DeepSkyBlue", "Black", "Fuchsia", "LightSkyBlue"]
labels = [  "Energia Cinetica.\n Media Prog",
            "Energia Potenziale.\nMedia Prog",
            "Energia Totale.\n Media Prog",
            "Pressione \nMedia Prog", 
            "Temperatura \n Media Blocchi",
            "Temperatura \n Media Prog"]
ylabels = ["K*/N", "U*/N", "E*/N", "p*", "T*", "T*"]
```
