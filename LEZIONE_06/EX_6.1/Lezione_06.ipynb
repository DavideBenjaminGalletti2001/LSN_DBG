<h3><font color = 'LightGreen'>Intro:</font></h3>

Di certo, uno dei modelli principe e maggiormente applicati in fisica statistica e' il cosiddetto modello di ising, il cui scopo
originario era quello di descrivere il comportamento magnetico dei materiali andando ad indagare eventuali fenomeni di transizione di fase nello stato ferromagnetico: una fase ordinata contrassegnata da un parametro d'ordine-rappresentato dalla magnetizzazione- diverso da zero in assenza di campo magnetico esterne ed originantesi per rottura di simmetria dovuta alla degenerazione rotta del GS provocata da una debole perturbazione esterna.
Oggigiorno, invece, viene fecondamente e frequentamente impiegato per descrivere fenomeni di transizioni di altri sistemi, come ad esempio le leghe binarie o le transizioni solido-liquido, le quali vengono mappate del modello di ising in modo tale da essere piu' facilmente studiati. Altre applicazioni notevoli possono essere trovate nei problemi di ottimizzazione, oppure altresi'nella progettazione dei computer quantistici.
D'altra parte, per quanto potente ed estremamente versatile, si conoscono il modello di Ising non e' sempre risolvibile in modo analitico: nel caso 1d vi e' la soluzione esatta sia in presenza che in assenza di campo magnetico esterno, al contrario, a dimensioni maggiori, il problema puo' diventare non 
banale; Onsager trovo' una soluzione analitica-cui ne sono seguite numerose varianti- del caso 2d in assenza di campo magnetico. In tutti gli altri casi invece occorre adottare un approccio numerico, anche se e' opportuno sottolineare che i casi di dimensione maggiore o uguale a quattro possono essere trattati efficacemente in approssimazione di campo medio portando a soluzioni tanto piu' precise quanto piu' cresce la dimensione. In fisica teorica-come in teoria delle stringhe- questo risulta essere particolarmente un risultato fortunato ed utile. Tuttavia tale approssimazione di campo medio fallisce significativamente a basse dimensione, come emerge in modo particolarmente limpido nel cas 1d dove non provede alcuna transizione di fase: cosa non vera e che viene dimostrata analiticamente; a tal riguardo, curioso e' notare che adottando proprio questo approccio questa fu la conclusione a cui giunse originariamente Ising stesso nella tesi di dottorato propostagli dal suo docente.
Come dicevo, la solubilita' di siffatto modello e' strettamente collegata alla dimensione in cui viene studiato e alla presenza o meno di un campo magnetico esterno. Per riassumere:
<lo>
    <li> 1d e $h = 0$: esatto</li>
    <li> 1d e $h \ne 0$: esatto</li>
    <li> 2d e $h = 0$: esatto(Onsager)</li>
    <li> 2d e $h \ne 0$: numerico</li>
    <li> nd, n$\geq 3$ e h$= \textbackslash $$\ne 0$: numerico</li>
</lo>


<h3><font color = 'LightGreen'>Esercitazione: Modello Ising</font></h3>

Entrando nello specifico, il compito di questa esercitazione e' studiare il caso 1d numericamente e sfruttare il risultato analitico per effettuare un confronto coi risultati ottenuti. Assumendo PBC in modo tale da intendere il sistema come un anello discreto, l'hamiltonia del modello di ising-in cui l'interazione e' assunta solo tra primi vicini- ha forma:
<p style = "text-align:center">
$H = -J\sum_{i=1}^N s_i s_{i+1}-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1})$
</p>
dove J rappresenta in termine di interazione(o di scambio) ed e' positivo nel caso ferromagnetico e negativo nel caso antiferromagnetico. Vivecerva, h indica il campo esterno.
Dalla teoria, si sa che la funzione di partizione di questo sistema e' della forma
<p style = "text-align:center">
$Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N$
</p>
dove $\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, $\beta=\frac{1}{k_b T} = \frac{1}{T}$<br>
Si noti che l'hp di condizioni periodiche introduce una simmetria in invarianza traslazionale.
La proceduta che si adopera e' di per se' relativamente semplice e si scandisce in pochi passaggi:
<lo>
    <li>Scegliere la <span style = "color:blue">configurazione iniziale</span>, la cui scelta non e' di per se' strettamente fondamentale da influenza 
        il raggiungimento all'equilibrio; essenzialmente si hanno due possibilita': spin random(T = $\infty$) o spin allineati(T = 0).</li>
    <li><span style = "color:blue">Equilibrazione</span>, per cui si intende il fatto che si sta campionando la corretta configurazione, cioe' che la 
        propabilita' di trovarci in un certo stato e' proprorzionale al peso di Boltzmann. Per implementarla, basta solo partire da una certa 
        configurazione iniziale e monitorare l'andamento di una particolare osservabile fintantoche' questa non saturera'. Per evitare di rimanere 
        bloccati in stati metastabile od neffienze nel rompere la struttura cristallina di partenza conviene testare differenti condizioni iniziale 
        questo lo si fa utilizzando per ognuna un seme differente per inizializzare il generatore di numeri pseudocasuali adoperato per effettuare la 
        simulazione.Possibili scelte di stati iniziali sono fase ordinata contro fase disordinata, oppure due stati degeneri della fase disordinata</li>
    <li>Fatta l'equilibrazione, si puo' principiare con la <span style = "color:blue">Simulazione</span> vera e propria in modo tale da ottenere la 
        termodinamica del sistema. Per la stima delle varie grandezze in gioco occorre utilizzare un metodo quale il data blocking o il piu' avanzato 
        resampling, i quali permettodo di risolvere il problema della correlazione dei campionamenti ottenuti. Nel data blocking, basta prendere blocchi 
        di dimensioni maggiore circa a due volte il temoo di autocorrelazione in mdo tale da assicurarsi che eventuali correlazioni siano solo meramente 
        effetti di superficie.</li>
</lo>

Per implementare la matrice di trasferimento e campionare il peso di boltzmann tramite la costruzione di un cammino di markov che stati tra stati a probabilita' invarianti, una possibilita' e' considerare come mossa tentativo una operazione casuale di flipping dello spin. Nel caso del Metropolis, si va a vedere se accettare o meno la mossa in base al comportamento dell'energia: se si finisce verso stati ad energie piu' bassa, questa viene accettatta con probabilita' $=1$; per converso, se si cresce di energia si avra' una probabilita' data dal peso relativo del peso di Boltzamnn. Per quanto concerno 
il campionamento con metodo di gibbs, sfruttando opportunamente le probabilita' condizionate in questo caso facilmente ricavabili, si puo' costruire una probabilita' di accettazione che e' sempre uguale alla certezza e quindi rendere particolarmente efficiente la simulazione.
Nello specifico, in questa esercitazione ci e' stato consegnato un codice gia' funzionamente in cui inserire il calcolo della capacita' termica, della suscettibilita' magnetica e della magnetizzazione. Quanto ci e' stato richiesto e' di determinare tali grandezze in funzione di $T \in \left[0.5;2.0\right]$. In particolare, si e' discretizzato l'intervallo in $T = \{0.5, 0.7, 1, 1.2, 1.4, 1.6, 1.8, 2.0\}$.<br>
In breve, le grandezze da determinare sono:<br>

<lo>
    <li>$U(N,T) = \langle H \rangle_{h=0}$</li>
    <li>$C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$</li> 
    <li>$\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$</li>
    <li>$M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$</li>
</lo>

Si ricordo, infine, che il reticolo 1d il PBC si traduce in un anello discreto e che si assume che questo e' costituito da $ N = 50$ siti.

<h3><font color = 'LightGreen'>Metropolis:</font></h3>

<lo>
    <li>prende configurazione $\nu = \{s_1, ..., s_n\}$</li>
    <li>Modifica orientazione spin $\mu = \{s_1, ..., -s_i, ..., s_n\}$</li>
    <li>accetta la mossa in base a $A(\mu|\nu) = \min \left[1, e^{-\beta(\Delta E)}\right]$</li>
</lo>

<h3><font color = 'LightGreen'>Gibbs:</font></h3>

l campionamento di Gibbs è nella pratica il più semplice campionamento lungo una catena di Markov. Ad ogni passaggio viene estratta una particella e ne si impone l'orientamento estraendone il valore con probabilità:
<p style = "text-align:center">
$T(\nu|\mu)= \frac{p(s_j'= \pm 1)}{p(s_j=+1)+p(s_j=-1)}=\frac{1}{1+ \text{exp}\left(- \beta \Delta E    \right)}$
</p>

<h2><font color = 'LimeGreen'>ESERCIZIO 6.1: Modello di ising, equilibrazione e termodinamica</font></h2>

<h3><font color = 'LightGreen'>Teoria: Modello Ising</font></h3>

Pe chiarezza di esposizione, in questo primo paragrafo si riportano le soluzione teoriche date dalla risoluzione esatta del modello di
ising 1d con condizioni periodiche al bordo:

```python
#INIZIALIZZAZIONE DATI PROBLEMA:

points=100
T = np.linspace(0.5,2,num=points)
beta = 1/T
J = 1.0 
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
print("DATI:")
print("PUNTI = ", points)
print("J =", J)
print("Ns =", Ns)
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")

```

<h3><font color = 'Black'>Energia interna</font></h3>

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
print("CALCOLO ANALITICO:")
plt.figure(figsize = (10,5), facecolor = "AliceBlue")
plt.plot(T, e,color = "DarkMagenta")

font = {'family':'serif', 'color':'forestgreen','size':'15'}
plt.ylabel('U/N', fontdict = font)
plt.xlabel('T', fontdict = font)
plt.grid(True)
plt.title('Ising 1D, internal energy', fontsize = 20, fontweight = "bold", color = "ForestGreen", family = "Serif")
plt.show()

print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
```

<h3><font color = 'Black'>Capacita' termica</font></h3>

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

```python
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.figure(figsize = (10,5), facecolor = "AliceBLue")
plt.plot(T, heat, color = "DarkMagenta")

font = {'family':'serif', 'color':'forestgreen','size':'15'}
plt.ylabel('C', fontdict = font)
plt.xlabel('T', fontdict = font)
plt.grid(True)
plt.title('Ising 1D, Heat capacity', fontsize = 20, fontweight = "bold", color = "ForestGreen", family = "Serif")
plt.show()
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
```

<h3><font color = 'Black'>Magnetizzazione $h \ne 0$</font></h3>

$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$

L'esatta formula da $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ e':<br> 
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

```python
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
h=0.02 #external field
b = 1/T
print("h = ", h)
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
```

```python


l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.figure(figsize = (10,5), facecolor = "AliceBlue")
plt.plot(T, M,color = "DarkMagenta")
font = {'family':'serif', 'color':'forestgreen','size':'15'}
plt.ylabel('M', fontdict = font)
plt.xlabel('T', fontdict = font)
plt.grid(True)
plt.title('Ising 1D, Heat capacity', fontsize = 20, fontweight = "bold", color = "ForestGreen", family = "Serif")
plt.show()
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
```

<h3><font color = 'Black'>Suscettibilita' $h = 0$ </font></h3>

$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.figure(figsize = (10,5), facecolor = "AliceBLue")
plt.plot(T, X, color = "DarkMagenta")

font = {'family':'serif', 'color':'forestgreen','size':'15'}
plt.ylabel('$\chi$', fontdict = font)
plt.xlabel('T', fontdict = font)
plt.grid(True)
plt.title('Ising 1D, Susceptibility', fontsize = 20, fontweight = "bold", color = "ForestGreen", family = "Serif")
plt.show()
print("")
print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
print("")
```

<h3><font color = 'LightGreen'>Equilibrazione:</font></h3>

Come anticipato, prima di qualsiasi misurazione vera e propria delle proprieta' termodinamiche del sistema, occorre eseguire la simulazione per un tempo macchina, quindi un certo numero di step, sufficiente affinche' raggiunga l'equilibrio alla temperatura desiderata. Durante l'inizializzazione, infatti, si e' scelto che la funzione Input setta ogni spin casualmente, cioe' nella condizione di temperatura infinita.
La funzione Move modifica le configurazioni in base alla loro temperatura, attraverso il parametro, quindi ci aspettiamo che dopo un certo numero di passi, il sistema raggiunga una condizione di stazionarieta' a cui la temperatura si assesta al valore di input.

Per andare ad osservare quando e dopo quanti passi il sistema raggiunge l'equilibrio, eseguiamo una simulazione con 1000 blocchi ognuno da uno step e analizziamo i valori istantanei step by step di energia interna e magnetizzazione per particella.
Il raggiungimento all'equilibrio si inferisce se queste grandezze andranno a saturare in un valore piu' o meno costante a meno di fluttuazioni statistiche accettabili: si noti l'originarsi di magnettizzazione a temperature basse.

```python
###############################################################
#EQUILIBRAZIONE: 

#ASCISSE:
N = 1000 #NUMERO BLOCCHI
x = np.arange(N) +1 #INTERVALLO ASCISSE: DA 1 A N = 1000
#IMPLEMENTAZIONE GRAFICO:
labels = ["ene", "mag"]
ylabels = ["ENERGIA", "MAGNETIZZAZIONE"]
titles = ["ENERGIA", "MAGNETIZZAZIONE"]
temp = ["0.5","1", "1.4"]
fields = ["0", "0.02"]
algorithms =["METROPOLIS","GIBBS"]
colors = ["Blue", "Indigo", "LightGreen"]
#***************************************************#
#CICLO SU METODO ADOPERATO: METROPOLIS E GIBBS.
#NEL CONCRETO, OTTENGO DUE FILE .jpeg UNO PER 
#METROPOLIS L'ALTRO PER GIBBS.
#IN CIASCUN FILE CI SONO 4 GRAFICI: I PRIMI
#DUE RELATIVI ALLA ENERGIA PER I DUE CAMPI 
#h = 0 E h = 0.02; I SECONDI ALLA MAGNETIZZAZIONE
#SEMPRE PER h = 0 E h = 0.02
#***************************************************#
for a in algorithms:
    plt.figure(figsize=(20,18), facecolor = "AliceBlue")
    if a == "METROPOLIS": plt.suptitle("METROPOLIS EQUILIBRAZIONE", fontsize = 25, c = "ForestGreen", fontweight = "bold", family = "serif") #TITOLO PRINCIPALE FILE METROPOLIS
    if a == "GIBBS": plt.suptitle("GIBBS EQUILIBRAZIONE", fontsize = 25, c = "ForestGreen", fontweight = "bold", family = "serif") #TITOLO PRINCIPALE FILE METROPOLIS
    conta = 0 
    appo = 0
    for l in labels:#CICLO SU OSSERVABILI EQUILIBRATE: ENERGIA E MAGNETIZZAZIONE
        for h in fields:#CICLO SUL NUMERO DI CAMPI h CONSIDERATI: IN PRATICA,OTTENGO L'ENERGIA PRIMA E LA MAGNETIZZAZIONE DOPO 
                        #PER I CAMPI h = 0 e h = 0.02
            conta+=1
            plt.subplot(4,1, conta)
            c = 0
            for t in temp:#PLOTTO SULLO STESSO GRAFICO LE TEMPERATURE DESIDERATE
                ave = np.loadtxt("EX_6.1_TOT/OUTPUT_FILE/EQUILIBRAZIONE/"
                                + a + "/ALL" +  "/h" + h + "/output."
                                + l + "_" + "h" +  h + "T" + t, usecols= (1),
                                unpack = "True")
                plt.plot(x, ave, marker = None, label = l + "h=" + h + "_T=" + t, alpha = 0.6, color = colors[c])
                c=c+1
            
            plt.title(titles[appo] + "_h=" + h, fontsize = 18, fontweight = "bold", c = "ForestGreen", family = "Serif")
            font = {"family":"serif", "color":"ForestGreen", "size":"15"}
            plt.ylabel(ylabels[appo], fontdict = font)
            plt.xlabel("step", fontdict = font)
            plt.grid(True)
            plt.legend()
            plt.subplots_adjust(hspace = 0.5)

        appo+=1
    title = "eq" + a + ".jpeg"
    plt.savefig(title)   
     
```

Nei qui presenti grafici, in gruppi da quattro, sono raffigurate l'equilibrazione nel caso Metropolis prima e Gibbs dopo: in ognuno dei casi, viene
plottato l'andamento dell'energia e magnetizzazione-prima in assenza e poi in presenza di campo esterno- per differenti temperature: interessante e' notare la formazione di megnettizzazione  e quindi la transizione di fase ferromagnetica nel caso di basse temperature.

<h3><font color = 'LightGreen'>Termodinamica: Modello Ising</font></h3>

Fatta l'equilibrazione e assicuratisi quindi di campionare la corretta distribuzione di probabilita', si puo' iniziare, partendo dalla configurazione equilibrata, la simulazione vera e propria per andare a studiare la termodinamica deldel sistema. Come usuale, al fine di eliminare la correlazione, tali misure istantanee sono state eleborate col metodo del data blocking: cosi' facendo, le singole grandezze indipendenti, su cui poi, tramite medie progressive, andare a calcolare l'incertezza statistica in conformita' con teoriema del limite centrale, non sono altro che le medie dei blocchi di campionamente tra loro correlati. Tali medie fluttueranno meno, riducendo l'effetto di correlazione, intrinseco nel metodo di metropolis per definizione
stessa di catena di Markov, ad un mero fenomeno di superficie tanto piu' trascurabile tanto maggiore e' la dimensione dei blocchi: una dimensione, questa, che si puo' determinare andando ad indagare l'autocorrelazione di osservabili fisiche del sistema e considerando come dimensione dei blocchi superiori a due volte il tempo di correlazione.
